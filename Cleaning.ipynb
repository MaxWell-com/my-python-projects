```python
total_sales = df.groupby('Номенклатура.Объединенная номенклатура.Код')[['01.03.2024', '01.04.2024', '01.05.2024',
                                                                         '01.06.2024', '01.07.2024', '01.08.2024',
                                                                         '01.09.2024', '01.10.2024', '01.11.2024',
                                                                         '01.12.2024', '01.01.2025', '01.02.2025']].sum()
```

```python
total_sales['Общая сумма Код'] = total_sales.sum(axis=1)
```

```python
df.head(5)
```

```python
df = df.merge(total_sales[['Общая сумма Код']], left_on='Номенклатура.Объединенная номенклатура.Код', right_index=True, how = 'left')
```
