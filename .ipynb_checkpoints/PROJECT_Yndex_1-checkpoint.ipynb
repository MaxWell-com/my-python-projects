```python
def triangle (a,b,c = 10):
    if not all(isinstance(x, int) for x in [a,b,c]):
        return 'Длина отрезков должна быть целым числом'
    if a <=0 or b <=0 or c <=0:
        return False

    return (a+b>c) and (a+c>b) and (b+c>a)

triangle(5,7,10)
triangle(1.3,-2,-3)
```

```python
# Можно начинать творить тут

#функция для расчета продукта-лидера по выручке за определенный день
#для того, чтобы эта функция сработала в датафрейме, к которому она применяется должны обязательно 
#присутствовать данные 'time_order', 'var_product', 'price', 'quantity'

import pandas as pd


def best_selling_product(df, date):
    df['time_order'] = pd.to_datetime(df['time_order'])
    target_date = pd.to_datetime(date).date()
    df_day = df[df['time_order'].dt.date == target_date]
    
    if df_day.empty:
        return None
    
    df_day = df_day.assign(revenue=df_day['price'] * df_day['quantity'])
        
    product_revenue = df_day.groupby('var_product')['revenue'].sum()
    
    best_product = product_revenue.idxmax()
    
    return print(f"Лучший продаваемый товар {date}: {best_product}")
    

#проверим корректность работы функции на нашем основном датафрейме 
best_selling_product(df, '2023-01-30')
```
